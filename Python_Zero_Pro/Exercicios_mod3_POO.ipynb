 Os métodos são os seguintes: alterarNome, depósito e saque; No construtor,  saldo é opcional, com valor default zero e os demais atributos são obrigatórios.

8. Classe Bomba de Combustível: Faça um programa completo utilizando classes e  métodos que: 


a. Possua uma classe chamada bombaCombustível, com no mínimo esses atributos: 

* i. tipoCombustivel. 
* ii. valorLitro 
* iii. quantidadeCombustivel 

b. Possua no mínimo esses métodos:
 
* i. abastecerPorValor( ) – método onde é informado o valor a ser abastecido e  mostra a quantidade de litros que foi colocada no veículo 
* ii. abastecerPorLitro( ) – método onde é informado a quantidade em litros de  combustível e mostra o valor a ser pago pelo cliente. 
* iii. alterarValor( ) – altera o valor do litro do combustível. 
* iv. alterarCombustivel( ) – altera o tipo do combustível. 
* v. alterarQuantidadeCombustivel( ) – altera a quantidade de combustível  restante na bomba. 

OBS: Sempre que acontecer um abastecimento é necessário atualizar a  quantidade de combustível total na bomba.


```python
class BombaDeCombustivel:
    def __init__(self,tipoCombustivel, valorLitro, quantidadeCombustivel):
        self.tipoCombustivel = tipoCombustivel
        self.valorLitro = valorLitro
        self.quantidadeCombustivel = quantidadeCombustivel
    
    
    def abastecerPorValor(self,valor):
        self.valor = valor
        self.abastecerValor = self.valor / self.valorLitro
        self.quantidadeCombustivel = self.quantidadeCombustivel - self.abastecerValor
        print(f'Abasteceu R${self.valor:.2f} e {self.abastecerValor:.2f} litros de {self.tipoCombustivel} | quantidade de litros na bomba {self.quantidadeCombustivel}')
       

    def abastecerPorLitro(self,litros):
        self.litros = litros
        self.abastecerLitros = self.litros * self.valorLitro
        self.quantidadeCombustivel = self.quantidadeCombustivel - self.litros
        print(f'Abasteceu R${self.abastecerLitros:.2f} e {self.litros:.2f} litros de {self.tipoCombustivel} | quantidade de litros na bomba {self.quantidadeCombustivel}')

        
    def alterarValor(self,novoValor):
        self.valorLitro = novoValor
                 
    def alterarCombustivel(self,novoCombustivel):
        self.tipoCombustivel = novoCombustivel
        

    def alterarQuantidadeCombustivel(self,novaQuantidade):
        self.quantidadeCombustivel = novaQuantidade
        



abastecer = BombaDeCombustivel('alcool',3.00, 500)
abastecer.abastecerPorValor(30)
abastecer.abastecerPorLitro(50)
abastecer.alterarValor(4)
abastecer.abastecerPorValor(30)
abastecer.abastecerPorLitro(50)
abastecer.alterarCombustivel('gasolina')
abastecer.abastecerPorValor(30)
abastecer.abastecerPorLitro(50)
abastecer.alterarQuantidadeCombustivel(600)
abastecer.abastecerPorValor(30)
abastecer.abastecerPorLitro(50)

```
